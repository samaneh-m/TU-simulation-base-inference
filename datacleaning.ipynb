```python
# Importación de librerías a utilizar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from numpy.core.function_base import linspace
from math import floor
from math import ceil
from numpy.ma.core import size
from sklearn.model_selection import train_test_split

# Lectura de la base de datos
filename = 'new cleaned data.csv'
data = pd.read_csv(filename, sep = ',')
```

```python
res = data[data['len'] <= 150].reset_index(drop=True)
```

```python
res = res.sample(frac=1).reset_index(drop=True)
```

```python
valid_aas = set(['A', 'R', 'N', 'D', 'C', 'E', 'Q', 'G', 'H', 'I',
                 'L', 'K', 'M', 'F', 'P', 'S', 'T', 'W', 'Y', 'V'])

def is_valid_sequence(seq):
    return all(aa in valid_aas for aa in seq)

# Apply the filter
df = res[res['seq'].apply(is_valid_sequence)].reset_index(drop=True)

df = df[~df['seq'].str.contains(r'\*', regex=True)].reset_index(drop=True)
```

```python
def convert_states_to_binary(seq):
    return ''.join(['H' if c == 'H' else 'O' for c in seq])

# Assuming your DataFrame is called `df` and the column is named 'state_seq'
binary_states = []

for i, row in res.iterrows():
    raw = row['sst3']  # Replace 'state_seq' with your actual column name
    binary = convert_states_to_binary(raw)
    binary_states.append(binary)

# Optionally add the result as a new column
# Rebuild the binary states for the filtered DataFrame
df['binary_state_seq'] = df['sst3'].apply(convert_states_to_binary)
```
